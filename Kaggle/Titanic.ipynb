```python
import pandas as pd
```

## Load Dataset

```python
train = pd.read_csv("Data/Titanic/train.csv", index_col=["PassengerId"])

print(train.shape)
train.head()
```

```python
test = pd.read_csv("Data/Titanic/test.csv", index_col=["PassengerId"])

print(test.shape)
test.head()
```

## Preprocessing

### Encode Sex
학습 알고리즘을 실행 시키기 위해 숫자형태로 바꾸어야 함

```python
train.loc[train["Sex"] == "male", "Sex"] = 0
train.loc[train["Sex"] == "female", "Sex"] = 1

print(train.shape)
train.head()
```

```python
test.loc[test["Sex"] == "male", "Sex"] = 0
test.loc[test["Sex"] == "female", "Sex"] = 1


print(test.shape)
test.head()
```

### Fill in missing fare
공백 값을 채움. 가장 기본적으로 평균으로 채우는 방법이 있다.
* test 데이터에만 공백이 존재하기 때문에 test만 채워줌

```python
mean_fare = test["Fare"].mean()
print("Fare(Mean) = ${0:.3f}".format(mean_fare))
```

```python

test.loc[pd.isnull(test["Fare"]), "Fare"] = mean_fare

test[pd.isnull(test["Fare"])]
```

### Encode Embarked
one hot encoding

```python
train["Embarked_C"] = train["Embarked"] == "C"
train["Embarked_S"] = train["Embarked"] == "S"
train["Embarked_Q"] = train["Embarked"] == "Q"

print(train.shape)
train[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
test["Embarked_C"] = test["Embarked"] == "C"
test["Embarked_S"] = test["Embarked"] == "S"
test["Embarked_Q"] = test["Embarked"] == "Q"

print(test.shape)
test[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

### Band Age
나이에 범위를 적용

```python
train["Age_Young"] = train["Age"] < 15
train["Age_Medium"] = (train["Age"] >= 15) & (train["Age"] < 35)
train["Age_Old"] = train["Age"] >= 35

print(train.shape)
train[["Age", "Age_Young", "Age_Medium", "Age_Old"]].head(10)
```

```python
test["Age_Young"] = test["Age"] < 15
test["Age_Medium"] = (test["Age"] >= 15) & (test["Age"] < 35)
test["Age_Old"] = test["Age"] >= 35

print(test.shape)
test[["Age", "Age_Young", "Age_Medium", "Age_Old"]].head(10)
```

## Train

```python
feature_names = ["Pclass", "Sex", "Fare",
                 "Embarked_C", "Embarked_Q", "Embarked_S","Age_Young","Age_Medium","Age_Old"]

X_train = train[feature_names]

print(X_train.shape)
X_train.head()
```

```python
label_name = "Survived"

y_train = train[label_name]

print(y_train.shape)
y_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

seed = 37

model = DecisionTreeClassifier(max_depth=5,
                               random_state=seed)
```

```python
model.fit(X_train, y_train)
```

## Predict

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head()
```

```python
prediction = model.predict(X_test)

print(prediction.shape)
prediction[:20]
```

## Submit

```python
submission = pd.read_csv("Data/Titanic/gender_submission.csv", index_col="PassengerId")

submission["Survived"] = prediction

print(submission.shape)
submission.head()
```

```python
submission.to_csv("Data/Titanic/submissions/faretest-script.csv")
```
